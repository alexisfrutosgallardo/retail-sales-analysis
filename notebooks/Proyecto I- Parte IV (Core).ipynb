```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('retail_sales_dataset.csv')
df = pd.DataFrame(data)
```

**Análisis Descriptivo y Visualización**
* Calcula estadísticas descriptivas básicas para comprender mejor las características de los datos.
* Utiliza histogramas y boxplots para visualizar la distribución de las variables numéricas clave.
* Crea gráficos de líneas para mostrar las tendencias de ventas a lo largo del tiempo.
* Utiliza gráficos de dispersión para analizar la relación entre diferentes variables.
* Combina histogramas y boxplots para visualizar la distribución de las ventas y detectar valores atípicos.

```python
# Calcula estadísticas descriptivas básicas para comprender mejor las características de los datos.
# Primeras 5 filas
print("Primeras 5 filas:")
print(df.head())

# Últimas 5 filas
print("\n\nÚltimas 5 filas:")
print(df.tail())

# Información del DataFrame
print("\n\nInformación del DataFrame:")
print(df.info())

# Tipo de datos de las columnas
print("\n\nTipo de datos de las columnas:")
print(df.dtypes)

# Cambiamos el tipo de dato de Date
df['Date'] = pd.to_datetime(df['Date'])

# Tipo de datos de las columnas
print("\n\nTipo de datos de las columnas:")
print(df.dtypes)

# Estadísticas descriptivas
print("\n\nEstadísticas descriptivas:")
print(df.describe())

# Campos vacíos
print("\n\nCampos vacíos:")
print(df.isnull().sum())

# Filas duplicadas
print("\n\nFilas duplicadas:")
print(df.duplicated().sum())

# Guardamos los datos en un archivo CSV
df.to_csv('retail_sales_dataset_clean.csv', index=False)
print("Datos guardados con éxito")
```

```python
# Utiliza histogramas y boxplots para visualizar la distribución de las variables numéricas clave
# Histograma de Edades
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=40, color='lightgreen', edgecolor='black')
plt.title('Distribución de variables numéricas (edades)')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
# Boxplot de Edades
plt.figure(figsize=(10, 6))
plt.boxplot(df['Age'], vert=False)
plt.title('Boxplot de Edades')
plt.xlabel('Edad')
plt.grid(True)
plt.show()
```

```python
# Histograma de Total Amount
plt.figure(figsize=(10, 6))
plt.hist(df['Total Amount'], bins=40, color='lightblue', edgecolor='black')
plt.title('Distribución de variables numéricas (Total Amount)')
plt.xlabel('Total Amount')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
# Boxplot de Total Amount
plt.figure(figsize=(10, 6))
plt.boxplot(df['Total Amount'], vert=False)
plt.title('Boxplot de Total Amount')
plt.xlabel('Total Amount')
plt.grid(True)
plt.show()
```

```python
# Crea gráficos de líneas para mostrar las tendencias de ventas a lo largo del tiempo.
plt.figure(figsize=(10, 6))

# Agrupa los datos por fecha y suma las ventas totales
date_sales = df.groupby('Date')['Total Amount'].sum()

plt.plot(date_sales.index, date_sales.values, marker='o', linestyle='-', color='blue')
plt.title('Tendencias de Ventas a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Monto Total')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Utiliza gráficos de dispersión para analizar la relación entre diferentes variables
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Total Amount'], alpha=0.5, color='green')
plt.title('Relación entre Edad y Monto Total')
plt.xlabel('Edad')
plt.ylabel('Monto Total')
plt.grid(True)
plt.show()
```

```python
# Combina histogramas y boxplots para visualizar la distribución de las ventas y detectar valores atípicos.
plt.figure(figsize=(10, 6))

# Subgráfico 1: Histograma
plt.subplot(1, 2, 1) # 1 fila, 2 columnas, posición 1
plt.hist(df['Total Amount'], bins=4, edgecolor='black', color='lightgreen')
plt.title('Distribución de Ventas')
plt.xlabel('Monto Total')
plt.ylabel('Frecuencia')
plt.grid(True)

# Subgráfico 2: Boxplot
plt.subplot(1, 2, 2) # 1 fila, 2 columnas, posición 2
plt.boxplot(df['Total Amount'], vert=False)
plt.title('Boxplot de Ventas')
plt.grid(True)

plt.tight_layout()
plt.show()
```
